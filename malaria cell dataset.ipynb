```python
from tensorflow.keras.models import Model
```

```python
inp=layers.Input(shape=(size,size,3))
```

# first conv2d layer

```python
con1=layers.Conv2D(32,kernel_size=(3,3),activation="relu",padding="same")(inp)
max1=layers.MaxPool2D((2,2))(con1)
nor1=layers.BatchNormalization(axis=-1)(max1)
drop1=layers.Dropout(0.20)(nor1)
```

# second conv2d layer

```python
con2=layers.Conv2D(32,kernel_size=(3,3),activation="relu",padding="same")(nor1)
max2=layers.MaxPool2D((2,2))(con2)
nor2=layers.BatchNormalization(axis=-1)(max2)
drop2=layers.Dropout(0.20)(nor2)
```

# flatten layers

```python
flat=layers.Flatten()(drop2)
```

# first dense layer

```python
hidd1=layers.Dense(512,activation="relu")(flat)
norm3=layers.BatchNormalization(axis=-1)(hidd1)
drop3=layers.Dropout(0.20)(norm3)
```

# second dense layer

```python
hidd2=layers.Dense(256,activation="relu")(drop3)
norm4=layers.BatchNormalization(axis=-1)(hidd2)
drop4=layers.Dropout(0.20)(norm4)
```

# output layer

```python
out=layers.Dense(2,activation="sigmoid")(drop4)
```

```python
mod=Model(inputs=inp,outputs=out)
```
